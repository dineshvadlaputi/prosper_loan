```python
# if any null row is there it will drop row
df1= dataset.dropna()
```

```python
df1.size
```

```python
# row numerical null values used median
median_val = dataset.median()
df2 = dataset.fillna(median_val)
```

```python
# caterigorical null vaues print
nalist=[]
for i in df2.columns:
    #print(i)
    nalist.append(str(i))
    if(df2[str(i)].isnull().sum()!=0):
        print(i)
        print(df2[str(i)].value_counts())
```

```python
#categorical null values convert to mode
for column in nalist:
    mode_value = df2[column].mode()[0]
    df2[column].fillna(mode_value,inplace=True)
```

```python
#checking null values
for i in df2.columns:
    #print(i)
    if(df2[str(i)].isnull().sum()!=0):
        print(i)
```

```python
#data cleaning is done
```

```python
#DATA ENCODING
#CONVERTING CATEGORICAL FEATURES INTO NUMERICAL
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()
```

```python
#checking bool values
for i in df2.select_dtypes(include='bool'):
    print(i)
    print(df2[i].value_counts())
```

```python
#ENCODING BOOL TO NUMERICAL
for i in df2.select_dtypes(include='bool'):
    df2[i]=label_encoder.fit_transform(df2[i])
```

```python
#CHECKING OBJECT VALUES
for i in df2.select_dtypes(include='object'):
    print(i)
    print(df2[str(i)].value_counts())
```

```python
df2['ListingCreationDate'] = pd.to_datetime(df2['ListingCreationDate'], format='%Y-%m-%d %H:%M:%S.%f')
df2['ClosedDate'] = pd.to_datetime(df2['ClosedDate'], format='%Y-%m-%d %H:%M:%S.%f')
df2['DateCreditPulled'] = pd.to_datetime(df2['DateCreditPulled'], format='%Y-%m-%d %H:%M:%S.%f')
df2['FirstRecordedCreditLine'] = pd.to_datetime(df2['FirstRecordedCreditLine'], format='%Y-%m-%d %H:%M:%S.%f')
df2['LoanOriginationDate'] = pd.to_datetime(df2['LoanOriginationDate'], format='%Y-%m-%d %H:%M:%S.%f')


```

```python
for i in df2.select_dtypes(include='object'):
    df2[i]=label_encoder.fit_transform(df2[i])
```

```python
def plotcat(colname):
    print(colname)
    print("after data cleaning")
    cate=df2[colname].value_counts()
    plt.bar(cate.index,cate)
    plt.show()
    print("before data cleaning")
    cate=dataset[colname].value_counts()
    plt.bar(cate.index,cate)
    plt.xticks(rotation=45)
    plt.figure(figsize=(40,6))
    plt.show()
```
